```python
import pandas as pd
```

```python
a = pd.read_parquet("transacciones.parquet")
```

```python
final_date = a['purchase_date'].max()
```

```python
one_week = pd.Timedelta(weeks=1)
final_date_day_week = final_date - one_week
new_transactions = a[a['purchase_date'] > final_date_day_week]
new_transactions.loc[:, 'purchase_date'] = new_transactions['purchase_date'] + one_week
```

```python
final_date_new = new_transactions['purchase_date'].max()
```
